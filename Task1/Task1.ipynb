### Функция для визуализации распределений стратегий

Функция draw(dist) принимает массив dist и визуализирует его в виде столбцовой диаграммы.

```python
def dr(dist, ax, color):
    x = np.arange(1, len(dist) + 1)
    ax.scatter(x, dist, color=color)  
    ax.vlines(x, 0, dist, color=color) 
    ax.set_xlim([0, len(dist) + 1])
    ax.set_ylim([0, np.max(dist) + 1])
    ax.grid(True, linestyle='--')  
```

### Функция для представления решения игры и визуализации стратегий игроков

Функция pr(matrix) принимает на вход матрицу игры и выводит решение игры (равновесие Нэша) с помощью функции nash_equilibrium.

```python
def pr(matrix):
    res = nash_equilibrium(matrix)
    print("Цена игры: ", res[2], end="\n\n") 
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))  
    print("Массив стратегии первого игрока: ", res[0])
    dr(res[0], axs[0], color='blue') 
    print("Массив стратегии второго игрока: ", res[1])
    dr(res[1], axs[1], color='green')  
    
    plt.show()
```

### 1. Спектр оптимальной стратегии состоит из одной точки (существует равновесие Нэша в чистых стратегиях)

```python
m = [[3, 1, 5], [1, 3, 3], [3, 3, 1]]
pr(m)
```

### 2. Спектр оптимальной стратегии неполон (некоторые чистые стратегии не используются)

```python
m = [[1, 2, 3], [4, 1, 2]]
pr(m)
```

### 3. Спектр оптимальной стратегии полон

```python
m = [[-1, 1], [1, -1]]
pr(m)
```
